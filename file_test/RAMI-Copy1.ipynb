```python
def colonne(dataframe):
    for i in range(len(dataframe)):
        test = False
        for col in dataframe.columns:
            if col == 'PM2.5_CF1_ug/m3':
                test = True
    return test
```

```python
df_centrale =[]
df_bovis = []
df_estaque = []
df_inter_mourepiane = []
df_calade = []
df_nerthe = []
df_pelouqe = []
df_mourepiane =[]
df_tuiles = []
df_standre = []
df_stlouis = []
df_protis =[]
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(f):
        if 'Bovis' in f:
            df = pd.read_csv(f)
            if colonne(df) == True:
                df_bovis.append(pd.read_csv(f))
        if 'Ecole' in f:
            df = pd.read_csv(f)
            if colonne(df) == True:
                df_centrale.append(pd.read_csv(f))        
        if 'ESTAQUE' in f:
            df = pd.read_csv(f)
            if colonne(df) == True:
                df_estaque.append(pd.read_csv(f))
        if 'INTERMARCHE' in f:
            df = pd.read_csv(f)
            if colonne(df) == True:
                df_inter_mourepiane.append(pd.read_csv(f))
        if 'CALADE' in f:
            df = pd.read_csv(f)
            if colonne(df) == True:
                df_calade.append(pd.read_csv(f))
        if 'NERTHE' in f:
            df = pd.read_csv(f)
            if colonne(df) == True:
                df_nerthe.append(pd.read_csv(f))
        if 'Pélouque' in f:
            df = pd.read_csv(f)
            if colonne(df) == True:
                df_pelouque.append(pd.read_csv(f))
        if 'MOUREPIANE capt' in f:
            df = pd.read_csv(f)
            if colonne(df) == True:
                df_mourepiane.append(pd.read_csv(f))
        if 'Fontaine' in f:
            df = pd.read_csv(f)
            if colonne(df) == True:
                df_tuiles.append(pd.read_csv(f))
        if 'André' in f:
            df = pd.read_csv(f)
            if colonne(df) == True:
                df_standre.append(pd.read_csv(f))
        if 'saint louis' in f:
            df = pd.read_csv(f)
            if colonne(df) == True:
                df_stlouis.append(pd.read_csv(f))
        if 'Protis' in f:
            df = pd.read_csv(f)
            if colonne(df) == True:
                df_protis.append(pd.read_csv(f))
```
